```python
# Create a DataFrame for the age group data
age_bar = pd.DataFrame({
    'Age Group': ['Children & Adolescents (0-17)', 'Young Adults (18-34)', 
                  'Middle-Aged Adults (35-64)', 'Older Adults (65 and Over)'],
    'Population': [978971, 1301147, 1833051, 607531]
})

# Plot the bar chart
ax = age_bar.plot.bar(x='Age Group', 
                       y='Population', 
                       title='Age Group Population in the City of Los Angeles',
                       color='red',
                       figsize=(10, 6))

# Customize the y-axis ticks
yticks = range(0, max(age_bar['Population']) + 200000, 200000)
ax.set_yticks(yticks)
ax.set_yticklabels([f'{int(tick / 1)}' for tick in yticks])

# Display the plot
import matplotlib.pyplot as plt
plt.show()
```

```python
import plotly.express as px
fig = px.bar(age_bar, 
             x='Age Group', 
             y='Population', 
             title='Age Group Population in the City of Los Angeles',
             labels={'Population': 'Population Count', 'Age Group': 'Age Categories'},
             color='Age Group',
             color_discrete_map={
                 'Children & Adolescents (0-17)': '#1f77b4',  # Blue
                 'Young Adults (18-34)': '#ff7f0e',  # Orange
                 'Middle-Aged Adults (35-64)': '#2ca02c',  # Green
                 'Older Adults (65 and Over)': '#d62728'  # Red
             }) 

# Set the background to transparent
fig.update_layout(
    plot_bgcolor='rgba(0,0,0,0)',  # Transparent background
    paper_bgcolor='rgba(0,0,0,0)'  # Transparent outer background
)

fig.update_traces(marker_line_width=0, textposition='outside')

fig.show()
```

## Examining SCAG High Injury Network

```python
# Upload HIN geojson dataset
HIN =gpd.read_file('data_midterm/High_Injury_Network_SCAG.geojson')
```

```python
# View HIN
HIN.info()
```

```python
# View headings
HIN.head(5)
```

```python
# Keep city of LA
HINla = HIN[HIN.CITY == 'Los Angeles']
```

```python
# Check for city of LA HIN
HINla.head(2)
```

```python
# Plot city of LA HIN map
HINla.plot()
```

```python
# Check and Convert CRS
la_age = la_age.to_crs('EPSG:4326')
print("Original CRS:", la_age.crs)
```

```python
# Conduct a Spatial Join
la_age_HIN = gpd.sjoin(la_age, HINla)
```

```python
# Check New Dataset
la_age_HIN.info()
```

```python
# Overlay a Map - Children & Adolescents (0-17)
import matplotlib.pyplot as plt

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the Children & Adolescents (0-17) population in LA
la_age.plot(ax=ax, column='Children & Adolescents (0-17)', legend=True, cmap='GnBu',
            legend_kwds={"label": "Children & Adolescents Population", "orientation": "vertical"})

# Overlay HIN on the City of LA (high-injury network in red)
HINla.plot(ax=ax, color='red', markersize=0.5)

# Remove axes and add a title
ax.axis('off')
ax.set_title('HIN over Children & Adolescents Population in LA, 2019')

# Save the map
plt.savefig('images_midterm/LA_HIN_Children_Adolescents.jpg')
plt.show()
```
